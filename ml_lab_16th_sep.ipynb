# **Gradient Descent**

**Importing libraries and creating a new data**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression
```

```python
from numpy.random.mtrand import random
X,y=make_regression(n_samples=200,n_features=1,n_informative=1,n_targets=1,random_state=11,noise=30)
```

```python
plt.scatter(X,y)
```

```python
from numpy._core.fromnumeric import shape
m_arr=np.linspace(-200,200,20)
b_arr=np.linspace(-200,200,20)
m_grid,b_grid=np.meshgrid(m_arr,b_arr) # This create a 2-D parameter space.

final=np.vstack((m_grid.ravel().reshape(1,400),b_grid.ravel().reshape(1,400))).T #This code flatten 20X20 into 1-D array of 400 values.

z_arr=[]
for i in range(final.shape[0]):
  z_arr.append(np.sum((y-final[i,0]*X.reshape(200)-final[i,1])**2))

z_arr=np.array(z_arr).reshape(20,20)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(x = m_arr, y = b_arr, z =z_arr)])

fig.update_layout(title='Cost Function', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))



fig.show()

fig.write_html("cost_function.html")
```

```python
print(X[0])
print(y[0])
```

```python
b=0
m=0
lr=0.01
all_b=[]
all_m=[]
all_cost=[]

epochs=30

for i in range (epochs):
  slope_m=0
  slope_b=0
  cost=0
  for j in range(X.shape[0]):
    slope_m=slope_m-2*(y[j]-(m*X[j])-b)*X[j]
    slope_b=slope_b-2*(y[j]-(m*X[j])-b)
    cost=cost+(y[j]-m*X[j]-b)**2
  b=b-lr*slope_b
  m=m-lr*slope_m
  all_b.append(b)
  all_m.append(m)
  all_cost.append(cost)
```

```python
print(all_b)
print(all_m)
print(all_cost)
```
